```python
# empty lists for storing the matches later 
mat1 = [] 
mat2 = [] 
p = [] 

# printing the pandas dataframes 
print("First dataframe:\n", df_rating_imdb, 
	"\nSecond dataframe:\n", df_wiki) 

# converting dataframe column to list 
# of elements 
# to do fuzzy matching 
list1 = df_rating_imdb['Title'].tolist() 
list2 = df_wiki['Manga series'].tolist() 

# taking the threshold as 82 
threshold = 90

# iterating through list1 to extract 
# it's closest match from list2 
for i in list1: 
	mat1.append(process.extract(i, list2, limit=df_rating_imdb.shape[0])) 
df_rating_imdb['matches'] = mat1 

# iterating through the closest matches 
# to filter out the maximum closest match 
for j in df_rating_imdb['matches']: 
	for k in j: 
		if k[1] >= threshold: 
			p.append(k[0]) 
	mat2.append(",".join(p)) 
	p = [] 


# storing the resultant matches back to dframe1 
df_rating_imdb['matches'] = mat2 
print("\nDataFrame after Fuzzy matching:") 
df_rating_imdb 
```

```python
df_rating_imdb.to_csv('rating_imdb_matched.csv', index=False)
```
