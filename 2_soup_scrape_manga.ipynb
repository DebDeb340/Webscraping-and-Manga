```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# URL de la page Wikipedia
url = 'https://en.wikipedia.org/wiki/List_of_best-selling_manga'

# Récupérer le contenu de la page
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser',from_encoding='utf-8')

# Trouver tous les tableaux dans la page ayant la classe 'wikitable'
tableaux = soup.find_all('table', class_='wikitable')

# Pour chaque tableau trouvé
for idx, tableau in enumerate(tableaux, start=1):
    # Convertir le tableau HTML en DataFrame en utilisant pandas
    # et attribuer un nom de variable dynamique à chaque DataFrame
    globals()[f"df{idx}"] = pd.read_html(str(tableau))[0]
    
    # Afficher le DataFrame avec des explications
    print(f"DataFrame pour le tableau {idx}:")
    print(globals()[f"df{idx}"])  # Afficher le DataFrame
    print("\n")  # Ajouter une ligne vide pour la lisibilité

# Ce code récupère chaque tableau de la page Wikipedia,
# le convertit en DataFrame à l'aide de pandas,
# puis affiche chaque DataFrame avec un commentaire indiquant le numéro du tableau.
```

```python
def ma_function_de_traitement(v): # v: str -> int
    import re
    #r permet de définir que l'on souhaite garder les valeurs avant le premier slash entre crochet avec un ou plusieurs caractères à l'intérieur
    r = r"\[.\]"
    v = str(v)#v est remplacé par str(v)
    result = re.sub(r, "", v) # on remplace ce qui match r par rien ("") dans v
    # entre crochets valeurs entre 0 et 9
    #le + signifie au moins 1 caract entre 0 et 9
    #/.* signifie le slash et les valeurs après celui-ci
    slash_paulette = r"([0-9]+)/.*"
    #on applique cette 
    result = re.sub(slash_paulette, "\1", result).replace('\x01','')
    return int(result) if result else 0

```

```python
#apply function to transform colum data as int
full_list['No. of collected volumes'] = full_list['No. of collected volumes'].map(ma_function_de_traitement)
```
